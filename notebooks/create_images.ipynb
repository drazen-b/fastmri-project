```python
import matplotlib.pyplot as plt

# Create the figure and axes
fig, axs = plt.subplots(2, 3, figsize=(10, 6))

# Flatten the axes array to iterate over the subplots
axs = axs.flatten()

# Loop through the subplots and plot the images
for i in range(6):
    # Load the image, replace this line with your own image loading code
    img = plt.imread(f'images/slices/{i+1}.png')  # Replace "image{i+1}.jpg" with the filename of your image

    # Plot the image on the current subplot
    axs[i].imshow(img)
    axs[i].axis('off')  # Turn off axis labels

    # Add label to the image
    axs[i].annotate(chr(97 + i) + ')', (0.05, 0.95), xycoords='axes fraction', ha='left', va='top', color='white', fontsize=12)
# Adjust spacing between subplots
plt.subplots_adjust(hspace=0.1, wspace=0.1)

# Show the plot
plt.show()
```

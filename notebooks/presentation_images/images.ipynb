```python
import matplotlib.pyplot as plt
from PIL import Image
import os

# Define the image paths
folders = ['AXFLAIR', 'AXT1', 'AXT1POST', 'AXT1PRE', 'AXT2']
recons = ['original', 'zf', 'cs', 'unet', 'varnet']

fig, axs = plt.subplots(len(folders), len(recons), figsize=(15, 15))

for i, folder in enumerate(folders):
    for j, recon in enumerate(recons):
        image_path = os.path.join(folder, recon + '.png')  # Construct the image path
        img = Image.open(image_path)  # Open the image file
        axs[i, j].imshow(img)  # Display the image
        axs[i, j].axis('off')  # Turn off the axis

        # If in the top row, set the column title
        if i == 0:
            axs[i, j].set_title(recon)
            
    # Set the row label on the left of the subplots
    fig.text(0.14, (len(folders)-i)/(len(folders)+1), folder, ha='center', va='center', rotation='vertical', fontsize=12)

plt.subplots_adjust(left=0.15, wspace=0.05, hspace=0.05)
plt.show()
```
